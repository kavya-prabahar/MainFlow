**Importing the necessary libraries:**

```python
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib.ticker import FuncFormatter
```

**Loading the dataset:**

```python
url = "https://raw.githubusercontent.com/kavya-prabahar/MainFlow/main/householdtask3.csv"
```

```python
df = pd.read_csv(url)
```

```python
df.drop_duplicates(inplace=True)
```

```python
df.info()
```

```python
years = df['year'].unique()
```

```python
# formatter for displaying numbers in millions with one decimal place

def millions(x, pos):
    'The two args are the value and tick position'
    return '%1.1fM' % (x * 1e-6)

formatter = FuncFormatter(millions)
```

**Data Visualization with Matplotlib**

```python
# visualising the number of ownership of households and washing machines by year using a bar chart

plt.figure(figsize=(10, 5))
plt.bar(df['year'] - 0.2, df['own'], width=0.4, label='Own')
plt.bar(df['year'] + 0.2, df['own_wm'], width=0.4, label='Own WM')
plt.xlabel('Year')
plt.ylabel('Number of Households')
plt.title('Ownership of Households and Washing Machines by Year')
plt.legend()
plt.xticks(years)
plt.gca().yaxis.set_major_formatter(formatter)
plt.show()
```

```python
# visualising the total no of households vs owned using a line chart

df_grouped = df.groupby('year').mean().reset_index()

plt.figure(figsize=(10, 5))
plt.plot(df_grouped['year'], df_grouped['tot_hhs'],color = "green", marker='o', label='Total Households')
plt.plot(df_grouped['year'], df_grouped['own'], color = "orange", marker='o', label='Own')
plt.xlabel('Year')
plt.ylabel('Number of Households')
plt.title('Average Number of Total Households and Owned Households by Year')
plt.legend()
```

```python
# visualising the average income and expenditure of houseoholds grouped by year using bar chart

plt.figure(figsize=(10, 5))
plt.bar(df['year'] - 0.2, df['income'], color = "red", width=0.4, label='Income')
plt.bar(df['year'] + 0.2, df['expenditure'], color = "blue", width=0.4, label='Expenditure')
plt.xlabel('Year')
plt.ylabel('Value')
plt.title('Income and Expenditure by Year')
plt.legend()
plt.xticks(years)
plt.show()
```
