**Loading the dataset:**

```python
import pandas as pd

#Reading the csv file from the github repository
url = 'https://raw.githubusercontent.com/kavya-prabahar/MainFlow/main/01.Data%20Cleaning%20and%20Preprocessing.csv'

#Reading the dataset as DataFrame
df = pd.read_csv(url)
```

```python
# Displaying the first few rows of the DataFrame
print(df.head())
```

```python
#Displaying the last few rows of te DataFrame
print(df.tail())
```

```python
#Understanding the shape of the dataset
print(df.shape)
```

```python
print(df.info())
```

**Handling Missing Values:**

```python
print("Features having missing values:")
print("\n")
df.select_dtypes(include=(['int64','float64'])).isna().sum()
```

```python
#Replacing missing values with its mean

print("After handling missing values")
print("\n")
df_num_col = df.select_dtypes(include=(['int64','float64'])).columns
for c in df_num_col:
  df[c].fillna(df[c].mean(),inplace=True)
df.select_dtypes(include=(['int64','float64'])).isna().sum()
```

**Calculating Summary Statistics**

```python
df.describe()
```

**Filtering data based on conditions:**

```python
filtered_df = df[(df['Y-Kappa'] > 25) & (df['BlowFlow'] < 1300)]

print(filtered_df)
```

```python
#shape of filtered DataFrame

filtered_df.shape
```

**Saving the filtered data to a new CSV file:**

```python
filtered_df.to_csv("Filtered_data.csv")
```
