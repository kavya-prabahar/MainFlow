```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
url = "https://raw.githubusercontent.com/kavya-prabahar/MainFlow/main/heart.csv"
data = pd.read_csv(url)
```

**1. People with and without heart disease:**

```python
heart_disease_counts = data.target.value_counts()
```

```python
values = [heart_disease_counts.get(1, 0),heart_disease_counts.get(0, 0)]

print("Number of people with heart disease (target = 1):", values[0])
print("Number of people without heart disease (target = 0):", values[1])
```

```python
data.target.value_counts().plot(kind = "bar", color = ["blue","green"])
plt.title("No. of people with heart disease")
plt.xlabel("Outcome")
plt.ylabel("Count")
plt.show()
```

**Inference made: 526 people has heart disease and 499 people doesn't have heart disease**

**2. People of which sex has most heart disease?**

```python
sex_ratio = pd.crosstab(data.sex, data.target)
print(sex_ratio)
```

```python
heart_disease_by_sex = data[data['target'] == 1]['sex'].value_counts()

print("Number of people with heart disease by sex:")
print("Male:", heart_disease_by_sex.get(1, 0))
print("Female:", heart_disease_by_sex.get(0, 0))
```

```python
labels = ['Female', 'Male']

counts = [heart_disease_by_sex.get(0, 0), heart_disease_by_sex.get(1, 0)]


plt.figure(figsize=(8, 6))
plt.bar(labels, counts, color=['orange', 'blue'])
plt.xlabel('Sex')
plt.ylabel('Number of People with Heart Disease')
plt.title('Heart Disease Cases by Sex')
plt.show()
```

**Inference Made: The majority of people with heart disease are male**

**3. People of which sex has which type of chestpain the most?**

```python
chest_pain_counts = data.groupby(['sex', 'cp']).size().unstack(fill_value=0)
print("Chest pain counts by gender:")
print(chest_pain_counts)
```

```python
chest_pain_counts.plot(kind='bar', figsize=(10, 7))
plt.xlabel('Sex (0 = Female, 1 = Male)')
plt.ylabel('Count')
plt.title('Chest Pain Types by Gender')
plt.legend(title='Chest Pain Type')
plt.show()
```

**Both the Genders has type 0 chest pain the most**

**4. People with which chest pain are most prone to heart disease?**

```python
chest_pain = pd.crosstab(data.cp,data.target[1])
print(chest_pain)
```

```python
sns.countplot(data = data, x = "cp", hue = "target")
plt.title("No. of people with chest pain")
plt.xlabel("Chest Pain")
plt.ylabel("Count")
plt.show()
```

**Inference made: People with type 0 chest pain are most prone to heart disease**

**5. Count of male and female with and without heart disease:**

```python
gender_disease_counts = data.groupby(['sex', 'target']).size().unstack(fill_value=0)
print(gender_disease_counts)
```

```python
sns.countplot(data = data, x = "target", hue = "sex")
plt.title("No. of males and females with heart disease")
plt.xlabel("Outcome")
plt.ylabel("Count")
plt.show()
```

**6. How does the exercise-induced angina (exang) differ between individuals with and without heart disease?',**

```python
exang_counts = data.groupby(['target', 'exang']).size().unstack(fill_value=0)

print("Count of exercise-induced angina by heart disease status:")
print(exang_counts)
```

```python
plt.figure(figsize=(10, 6))
sns.countplot(data=data, x='target', hue='exang')
plt.title('Exercise-Induced Angina by Heart Disease Status')
plt.xlabel('Heart Disease Status (0 = No, 1 = Yes)')
plt.ylabel('Count')
plt.legend(title='Exercise-Induced Angina', labels=['No', 'Yes'])
plt.show()
```

**Inference Made: When Exercise-Induced Angina is true, heart diease is false and vice versa**

**7. Distribution of ages among males and females in the dataset:**

```python
print("Statistical Summary of Ages by Gender:")
print(data.groupby('sex')['age'].describe())
```

```python
plt.figure(figsize=(14, 6))
sns.histplot(data=data, x='age', hue='sex', multiple='stack', bins=30, palette='husl')
plt.title('Age Distribution by Gender (Histogram)')
plt.xlabel('Age')
plt.ylabel('Count')
```

**8. Distribution of resting blood pressure among individuals with and without heart disease:**

```python
print("Statistical Summary of Resting Blood Pressure by Heart Disease Status:")
print(data.groupby('target')['trestbps'].describe())
```

```python
plt.figure(figsize=(14, 6))
sns.histplot(data=data, x='trestbps', hue='target', multiple='stack', bins=30, palette='husl')
plt.title('Resting Blood Pressure Distribution by Heart Disease Status (Histogram)')
plt.xlabel('Resting Blood Pressure (trestbps)')
plt.ylabel('Count')
```

**Inference Made: People with Resting Blood Pressure 120 and 130 experiences the heart disease the most**

**9. Average age of the individuals in the dataset:**

```python
average_age = data['age'].mean()
print(f"The average age of individuals in the dataset is: {average_age:.2f}")
```

**10. Average cholesterol level for individuals with heart disease compared to those without:**

```python
average_chol = data.groupby('target')['chol'].mean()
print("Average Cholesterol Levels:")
print(f"Without heart disease (target = 0): {average_chol[0]:.2f}")
print(f"With heart disease (target = 1): {average_chol[1]:.2f}")
```
